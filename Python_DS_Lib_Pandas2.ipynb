#Chiporders

```python
import pandas as pd
df = pd.read_csv('http://bit.ly/chiporders',sep='\t')
df.head()
```

```python
df.dtypes
```

```python
#.str.replace() - to replace a charater in a data column
df['item_price'] = df['item_price'].str.replace('$','')
```

```python
df.head()
```

```python
df.dtypes
```

```python
#.astype() - to change the type of data in a data column
df['item_price'] = df['item_price'].astype(float)
```

```python
#Average price value of all the items present
df['item_price'].mean()
```

```python
#.str.contains() - verify the presence of any character or word in a data column
(df['item_name'].str.contains('Chips')).sum()
```

```python
dfc = df[df['item_name'].str.contains('Chicken')]
dfc.head()
```

```python
#Sum of total amount collected from each item
df.groupby('item_name')['item_price'].sum()
```

```python
#Average item price of each items ordered...
df.groupby('item_name')['item_price'].sum() / df.groupby('item_name')['quantity'].sum()
```

```python
df.groupby('item_name')['quantity'].sum()
```

#Drinks by Country

```python
df = pd.read_csv('http://bit.ly/drinksbycountry')
df.head()
```

```python
#Maximum quantity out of three liquors
df['Max Quantity'] = df[['beer_servings','spirit_servings','wine_servings']].max(axis=1)
```

```python
df.head()
```

```python
#Type of maximum quantity served out of three
df['Max_Type'] = df[['beer_servings','spirit_servings','wine_servings']].idxmax(axis=1)
```

```python
df.head()
```

```python
#Give the count of each sprit in max type column
#plot the continent wise serving of each type of liqour
```

```python
df['Max_Type'].value_counts()
```

```python
df['beer_servings'].sum()
```

```python
df.head()
```

```python
print(df[df['continent'] == 'Africa']['beer_servings'].sum())
print(df[df['continent'] == 'Africa']['spirit_servings'].sum())
print(df[df['continent'] == 'Africa']['wine_servings'].sum())
```

```python
#Total servings in each continent
df.groupby('continent')['beer_servings','spirit_servings','wine_servings'].sum()
```

```python
df.groupby('continent')['beer_servings','spirit_servings','wine_servings'].sum().plot.bar()
```

```python
#Average servings in each continent
df.groupby('continent')['beer_servings','spirit_servings','wine_servings'].mean()
```
