```python
#groupby() - function to group the data based upon some categorical column information
df2.groupby('Pclass')['Age'].mean()
```

```python
#Fill the null or void spaces - Mean or Mode values
df_na = df2[df2['Age'].isnull()]      #data samples with null values in age column
df_ma = df2[~df2['Age'].isnull()]     #data Samples with no null values in age column
```

```python
def fill_null_age(x):
  if x == 3:
    return 38.10
  elif x == 2:
    return 29.88
  else:
    return 25.14
```

```python
fill_null_age(1)
```

```python
df_na.head()
```

```python
#Applying any function created by us on a data column
df_na['Age'] = df_na['Pclass'].apply(fill_null_age)
```

```python
df_ma.head()
```

```python
df3 = pd.concat((df_na,df_ma),ignore_index=True)
df3.head()
```
